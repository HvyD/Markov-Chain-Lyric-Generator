```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open('data/all_artist_lyrics.txt')
Lyrics = f.read()
Lyrics = Lyrics.split()
df = pd.DataFrame(Lyrics)
```

```python
df
```

```python
wordcloud = WordCloud(max_font_size=42).generate(str(df))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
